```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Definir las variables iniciales
mean1 = 55
std_dev1 = 10
num_samples = 500

# Generar los números para la primera columna
column1_numbers = np.random.normal(mean1, std_dev1, num_samples)
column1_numbers = np.clip(column1_numbers, 30, 120)  # Limitar valores entre 30 y 120
column1_numbers = np.round(column1_numbers).astype(int)

# Definir las variables para la segunda columna
mean2 = 18
std_dev2 = 3

# Generar los números para la segunda columna
column2_numbers = np.random.normal(mean2, std_dev2, num_samples)
column2_numbers = np.clip(column2_numbers, 12, 26)  # Limitar valores entre 12 y 26
column2_numbers = np.round(column2_numbers).astype(int)

# Generar los números para la tercera columna
column3_numbers = np.random.randint(2, size=num_samples)

# Cambiar valores de la tercera columna basados en la primera columna
column3_numbers[column1_numbers > mean1] = 1

# Crear el DataFrame
data = {
    'Miles_Per_Week': column1_numbers,
    'Farthest_Run': column2_numbers,
    'Qualified_Boston_Marathon': column3_numbers
}

df = pd.DataFrame(data)
```

```python
df
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Miles_Per_Week'], df['Farthest_Run'], c=df['Qualified_Boston_Marathon'], cmap='viridis')

# Añadir etiquetas y título
plt.xlabel('Miles Per Week')
plt.ylabel('Farthest Run')
plt.title('Miles per Week vs Farthest Run (Qualified for Boston Marathon)')

# Mostrar el gráfico
plt.colorbar(label='Qualified for Boston Marathon')
plt.show()
```

```python
X = df.iloc[:, 0:2]
```

```python
y= df.iloc[:,2]
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=33,test_size=0.2)
```

```python
 from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
#Regularizacion
model_reg0 = SVC(C=0.1)
```

```python
model_reg0.fit(X_train, y_train)
```

```python
model_reg0.score(X_test, y_test)
```

```python
model_reg1= SVC(C=1)
model_reg1.fit(X_train, y_train)
model_reg1.score(X_test, y_test)
```

```python
#GAMMA
```

```python
model_gamma1= SVC(gamma=1.005)
model_gamma1.fit(X_train, y_train)
model_gamma1.score(X_test, y_test)
```

```python
#Kernel
model_linear = SVC(kernel='linear')
model_linear.fit(X_train, y_train)
model_linear.score(X_test, y_test)
```
