```python
import pandas as pd
```

```python
import numpy as np
```

```python
# Definir los parámetros
mean1 = 55
std_dev1 = 10
num_samples = 500

# Generar los números para la primera columna (Miles Per Week)
column1_numbers = np.random.normal(mean1, std_dev1, num_samples)
column1_numbers = np.clip(column1_numbers, 30, 120)  # Limitar valores entre 30 y 120
column1_numbers = np.round(column1_numbers).astype(int)

# Definir las variables para la segunda columna (Farthest Run)
mean2 = 18
std_dev2 = 3

# Generar los números para la segunda columna
column2_numbers = np.random.normal(mean2, std_dev2, num_samples)
column2_numbers = np.clip(column2_numbers, 12, 26)  # Limitar valores entre 12 y 26
column2_numbers = np.round(column2_numbers).astype(int)

# Generar los números para la tercera columna (Qualified for Boston Marathon)
column3_numbers = np.random.randint(2, size=num_samples)
column3_numbers[column1_numbers > mean1] = 1  # Si las millas por semana son mayores que la media, califican

# Crear el DataFrame
data = {
    'Miles_Per_Week': column1_numbers,
    'Farthest_Run': column2_numbers,
    'Qualified_Boston_Marathon': column3_numbers
}

df = pd.DataFrame(data)
```

```python
df.head(10)
```

```python
df = df.drop(columns=['Age','Gender'])
```

```python
df.head(10)
```

```python
X = df.iloc[:,0:2]
```

```python
y= df.iloc[:,2]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=26)
```

```python
#Hacemos un random forest, con arboles binarios (mayor cantidad de votos, la eleccion)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
#Vamos a usar un parameter grid, leer scikit documentation
```

```python
param_grid = [{
    'n_estimators':[500,1000,1500],
    #'criterion': ['entropy','gini'],
    'min_samples_split':[5,10,15],
    'min_samples_leaf': [1,2,4],
    #'max_depth':[10,20,30]
}]
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(rf,
                           param_grid,
                           cv=2, #multiplied by 2
                           scoring='accuracy',
                           n_jobs=-1
                           
    

    
)
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_param_grid = [{
    'n_estimators':[500,1000,1500],
    'criterion': ['entropy','gini'],
    'min_samples_split':[5,10,15],
    'min_samples_leaf': [1,2,4],
    'max_depth':[10,20,30]
}]
```

```python
random_grid_search = RandomizedSearchCV(rf,
                                        random_param_grid,
                                        cv=5,
                                        scoring='accuracy',
                                        n_jobs= -1,
                                        random_state=26
)
```

```python
random_grid_search.fit(X_train,y_train)
```

```python
random_grid_search.best_score_
```
