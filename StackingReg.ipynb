```python
import pandas as pd 
```

```python
import seaborn as sns 
```

```python
mpg = sns.load_dataset("mpg")
```

```python
mpg.head()
```

```python
mpg.head(30)
```

```python
mpg = mpg.drop('name',axis=1)
```

```python
mpg = pd.get_dummies(mpg)
```

```python
mpg.head()
```

```python
pd.DataFrame(mpg.isnull().sum().sort_values(ascending=False))
```

```python
mpg['horsepower'].fillna(mpg['horsepower'].mean(),inplace=True)
```

```python
pd.DataFrame(mpg.isnull().sum().sort_values(ascending=False))
```

```python
X = mpg.drop('mpg',axis=1)
```

```python
y = mpg['mpg']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,random_state=19)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
 from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(random_state=13)
```

```python
rfr.fit(X_train,y_train)
```

```python
y_pred2 = rfr.predict(X_test)
```

```python
r2_score(y_test,y_pred2)
```

```python
mean_squared_error(y_test,y_pred2)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge()
```

```python
param_grid = {
    'alpha': [0.1,0.3,0.5,0.7,0.9,1,3,5,10,15,30,50,75,500]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
ridge_cv = GridSearchCV(ridge, param_grid, cv=5, n_jobs=-1, scoring='neg_mean_squared_error')
```

```python
ridge_cv.fit(X_train,y_train)
```

```python
y_pred3 = ridge_cv.predict(X_test)
```

```python
r2_score(y_test,y_pred3)
```

```python
mean_squared_error(y_test,y_pred3)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr = GradientBoostingRegressor()
```

```python
gbr.fit(X_train,y_train)
```

```python
y_pred4 = gbr.predict(X_test)
```

```python
r2_score(y_test, y_pred4)
```

```python
mean_squared_error(y_test, y_pred4)
```

```python
from sklearn.ensemble import StackingRegressor
```

```python
estimators2 = [
    ('rfr', rfr),
    ('ridge',ridge_cv.best_estimator_),
    ('gbr',gbr)
]
```

```python
sr = StackingRegressor(
    estimators = estimators2,
    final_estimator = gbr
)
```

```python
sr.fit(X_train, y_train)
```

```python
y_pred5 = sr.predict(X_test)
```

```python
r2_score(y_test, y_pred5)
```

```python
mean_squared_error(y_test,y_pred5)
```

```python
from sklearn.svm import SVR
```

```python
estimators3= [
    ('ridge',ridge_cv.best_estimator_),
    ('gbr',gbr),
    ('svr',SVR(C=1.0, kernel='linear')),
    ('random_forest',RandomForestRegressor())
]
```
