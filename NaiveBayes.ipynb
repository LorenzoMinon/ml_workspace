```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)  # Para reproducibilidad

city_population = np.random.randint(1000000, 10000000, 500)  # Población de la ciudad entre 1M y 10M
continent = np.random.choice(['North America', 'Europe', 'Asia', 'South America', 'Africa', 'Australia'], 500)
venue_capacity = np.random.randint(5000, 50000, 500)  # Capacidad de los lugares entre 5K y 50K
day_of_week = np.random.choice(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], 500)
multiple_concerts = np.random.randint(0, 2, 500)  # 0 o 1 si hubo múltiples conciertos
sold_out = np.random.randint(0, 2, 500)  # 0 o 1 si se vendió todo

# Crear el DataFrame
data = {
    'City Population': city_population,
    'Continent': continent,
    'Venue Capacity': venue_capacity,
    'Day of Week': day_of_week,
    'Multiple Concerts': multiple_concerts,
    'Sold Out': sold_out
}

df = pd.DataFrame(data)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,y_pred), 
      (confusion_matrix(y_test,y_pred)))
```
