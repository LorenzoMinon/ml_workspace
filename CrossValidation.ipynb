```python
import pandas as pd
```

```python
import numpy as np
```

```python
np.random.seed(42)
```

```python
fastball_speed = np.random.randint(90,106, size =500)
```

```python
tommy_john = np.where(fastball_speed > 96, np.random.choice([0,1], size=500, p=[0.3, 0.7]), 0)
```

```python
d = {'fastball_speed':fastball_speed, 'tommy_john': tommy_john}
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
X = df[['fastball_speed']]
```

```python
y = df[['tommy_john']]
```

```python
#normal train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=11)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train.values.ravel())
```

```python
lr.score(X_test, y_test)
```

```python
#Test again
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X,y, test_size=0.2, random_state=25)
```

```python
lr.fit(X2_train,y2_train.values.ravel())
```

```python
lr.score(X2_test, y2_test)
```

```python
#Basic cross val
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cvscore = cross_val_score(lr, X,y.values.ravel(),cv=10) #We'll get 10 results from this
```

```python
cvscore
```

```python
np.average(cvscore)
```

```python
np.std(cvscore)
```

```python
#kfold example, define standard kfold
```

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits = 15, shuffle = True, random_state=42) #How many splits??
```

```python
kfscore = cross_val_score(lr, X, y.values.ravel(), cv=kf, scoring='f1')
```

```python
kfscore2 = cross_val_score(lr, X, y.values.ravel(), cv=kf, scoring='accuracy')
```

```python
kfscore
```

```python
np.average(kfscore)
```

```python
kfscore2
```

```python
np.average(kfscore2)
```

```python
np.std
```

```python
np.std(kfscore)
```

```python
np.std(kfscore2)
```

```python
#Stratified k fold
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
kf3 = StratifiedKFold(n_splits=10, shuffle= True, random_state=11)
```

```python
kfscore3 = cross_val_score(lr,X,y.values.ravel(), cv = kf3)
```

```python
kfscore3
```

```python
kfscore3.average()
```

```python
np.average(kfscore3)
```

```python
np.std(kfscore3)
```

```python
#pipeline example
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
pipe1 = make_pipeline(scaler,lr)
```

```python
scorespipe = cross_val_score(pipe1, X, y.values.ravel(), cv = 10)
```
