```python
"""Regularización L2: A diferencia de la regularización L1 (Lasso), que penaliza la suma de los valores absolutos de los coeficientes, la regularización L2 penaliza la suma de los cuadrados de los coeficientes. Esto significa que se agrega un término a la función de costo que es proporcional al cuadrado de la magnitud de los coeficientes."""
```

```python
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=100, n_features=4,noise=1,random_state=42,effective_rank=2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=19)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge()
```

```python
ridge.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
y_pred = ridge.predict(X_test)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
param_grid = {
    'alpha': [0.001,0.01,0.1,1,10,100]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
ridge_cv = GridSearchCV(ridge, param_grid, cv=3, n_jobs=-1)
```

```python
ridge_cv.fit(X_train,y_train)
```

```python
y_pred2 = ridge_cv.predict(X_test)
```

```python
mean_absolute_error(y_test,y_pred2)
```

```python
mean_squared_error(y_test,y_pred2)
```

```python
r2_score(y_test,y_pred2)
```
